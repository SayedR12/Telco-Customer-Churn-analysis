```python
df["TotalCharges"] = pd.to_numeric(
    df["TotalCharges"],
    errors="coerce"
)
```

```python
df["TotalCharges"] = df["TotalCharges"].fillna(0)
```

```python
#remove distanceds from object cilumns
for col in df.select_dtypes(include="object"):
    df[col] = df[col].str.strip()
```

```python
for col in df.columns:
    print(col)
    print(df[col].unique())
    print("-"*50)
```

```python
replace_cols = [
    "OnlineSecurity",
    "OnlineBackup",
    "DeviceProtection",
    "TechSupport",
    "StreamingTV",
    "StreamingMovies"
]

for col in replace_cols:
    df[col] = df[col].replace(
        "No internet service",
        "No"
    )
```

```python
replace_cols = [

    "MultipleLines"
]

for col in replace_cols:
    df[col] = df[col].replace(
        "No phone service",
        "No"
    )
```

```python
binary_cols = [
    "Partner",
    "Dependents",
    "PhoneService",
    "PaperlessBilling",
    "Churn"
]

for col in binary_cols:
    df[col] = df[col].map({
        "Yes":1,
        "No":0
    })
```

```python
Q1 = df["MonthlyCharges"].quantile(0.25)
Q3 = df["MonthlyCharges"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[
    (df["MonthlyCharges"] < lower) |
    (df["MonthlyCharges"] > upper)
]
```

```python
df["TenureGroup"] = pd.cut(
    df["tenure"],
    bins=[0,12,24,48,72],
    labels=[
        "0-12 Months",
        "13-24 Months",
        "25-48 Months",
        "49-72 Months"
    ]
)
```

```python
df["ChargeCategory"] = pd.qcut(
    df["MonthlyCharges"],
    q=3,
    labels=[
        "Low",
        "Medium",
        "High"
    ]
)
```

```python
df["CLV"] = (
    df["MonthlyCharges"]
    *
    df["tenure"]
)
```

```python
df.to_csv(
    "telco_cleaned.csv",
    index=False
)
```
